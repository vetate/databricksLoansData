```
%python
import pyspark.sql.functions as F

# Read the loan_analysiss table into a DataFrame
loan_df = spark.read.table("loan_analysiss")

# Group by Loan Category, calculate the sum of Loan, and order by descending sum
grouped_df = loan_df.groupBy("Loan Category").agg(F.sum("Loan").alias("Total_Loan")).orderBy(F.desc("Total_Loan"))

# Display all rows of the grouped DataFrame
display(grouped_df)
```

```
%python
# Read the loan_analysiss table into a DataFrame
loan_df = spark.read.table("loan_analysiss")

# Group by Marital Status, calculate the sum of Loan, and order by Marital Status
grouped_df = loan_df.groupBy("Marital Status").agg(F.sum("Loan")).orderBy("Marital Status")

# Display all rows of the grouped DataFrame
display(grouped_df)
```
